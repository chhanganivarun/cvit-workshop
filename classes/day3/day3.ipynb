**Origin and dialation/erosion**
1. convlove the stucturing element over the image
2. Dialation:
    - If any cell of structuring element overlaps with a black pixel of original image, set the origin to label black.
    Erosion:
    - If all cell of structuring element overlaps with a black pixel of original image, set the origin to label black.

```python

I = cv2.imread(path+'FingerPrint.jpg', cv2.IMREAD_GRAYSCALE)
ret,binary = cv2.threshold(I,125,255,cv2.THRESH_BINARY_INV)
se = morp.square(3,dtype='uint8')
print (se)
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, se)
closing = cv2.morphologyEx(opening, cv2.MORPH_CLOSE, se)
closing2 = closing/255
final = morp.medial_axis(closing2)*255
plt.imshow(np.hstack([I, binary, opening, closing, final]),cmap='gray')
```

```python
%matplotlib inline
I = cv2.imread(path+'FingerPrint.jpg', cv2.IMREAD_GRAYSCALE)

ret,I = cv2.threshold(I,125,255,cv2.THRESH_BINARY_INV)
ret, I_inv =cv2.threshold(I,125,255,cv2.THRESH_BINARY_INV)

b1 = np.array([[0,1,0],[1,0,1],[0,1,0]],dtype='uint8')
b2 = np.array([[0,0,0],[0,1,0],[0,0,0]],dtype='uint8')

erosion1 = cv2.erode(I,b2,iterations = 1)
erosion2 = cv2.erode(I_inv,b1,iterations = 1)

final = np.logical_and(erosion1,erosion2)*255

plt.imshow(np.hstack([I,erosion1,erosion2,final]),cmap = 'gray')
```
